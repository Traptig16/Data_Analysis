```python
#5
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("D:\\Internship\\matplotlib\\superstore1.csv")
df['Discount'] = df['Discount'].str.replace('%', '').astype(float)
average_discount = df.groupby('Region')['Discount'].mean()
average_discount.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Discount')
plt.title('Average Discount Across Regions')
plt.show()
highest_average_discount_region = average_discount.idxmax()
print("The region with the highest average discount is:", highest_average_discount_region)
print("Comment:Central Region gives maximum Discount")





```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("D:\\Internship\\matplotlib\\superstore1.csv")
average_profit = df.groupby('Region')['Profit'].mean()
average_profit.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.title('Average Profit Across Regions')
plt.show()
print("Comment:West Region gains maximum profit by giving minimum discount while Central region gives maximum discount then also it is in mimimum profit area. ")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

west_df = df[df['Region'] == 'West']

ship_mode_counts = west_df['Ship Mode'].value_counts()

ship_mode_counts.plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Profit from West Region')
plt.title('Ship Mode Distribution in the West Region')
plt.show()
print("Comment:West Region People prefferd standard mode")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
profit_by_ship_mode = df.groupby('Ship Mode')['Profit'].mean()
profit_by_ship_mode.plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Average Profit')
plt.title('Average Profit by Ship Mode')
plt.show()
print("comment:We gained maximum profit from first class")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.title('Total Sales by Region')
plt.show()
print("Comment:We achieved maximum sales and profit from west region by giving minimum discount but the west region people preffered standard class and we achieve least profit from standard class and gain highest profit from first class.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
ship_mode_count_by_region = df.groupby(['Region', 'Ship Mode']).size().unstack()
ship_mode_count_by_region.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Ship Mode Distribution by Region')
plt.legend(title='Ship Mode', loc='upper right')
plt.show()
print("comment: From this graph i conclude that most of the people preffered standard class compares to first class and maximum people are from west region so the profitable region is WEST")

```
