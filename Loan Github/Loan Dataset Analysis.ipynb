```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("D:\\Internship\\loan\\loan.csv")
df = df[['int_rate', 'grade']]
df['int_rate'] = pd.to_numeric(df['int_rate'].str.strip('%'))
grouped_data = df.groupby('grade')['int_rate'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['grade'], grouped_data['int_rate'])
plt.xlabel('Grade')
plt.ylabel('Mean Interest Rate')
plt.title('Mean Interest Rate by Grade')
plt.show()
print("G grade has highest interst rate")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("D:\\Internship\\loan\\loan.csv")
df = df[df['sub_grade'].str.startswith('B')][['int_rate', 'sub_grade']]
df['int_rate'] = pd.to_numeric(df['int_rate'].str.strip('%'))
colors = ['blue', 'orange', 'green', 'red', 'purple']
plt.figure(figsize=(10, 6))
for i, subgrade in enumerate(df['sub_grade'].unique()):
    subgrade_data = df[df['sub_grade'] == subgrade]
    plt.scatter(subgrade_data['sub_grade'], subgrade_data['int_rate'], color=colors[i], alpha=0.7, label=subgrade)

plt.xlabel('Subgrade')
plt.ylabel('Interest Rate')
plt.title('Interest Rate vs Subgrade (Starting with B)')
plt.legend()
plt.show()
print("The mostly purpose of the loan is debt consolidation which is taken by B grade people in which B5 has highest intrest rate")
```

```python
import pandas as pd
df=pd.read_csv("D:\\Internship\\loan\\loan.csv")
status_counts = df['loan_status'].value_counts()
total_loans = status_counts.sum()
risky_loans = status_counts[['Current', 'Charged Off']].sum()
risk_percentage = (risky_loans / total_loans) * 100
print(f"Risk Percentage: {risk_percentage:.2f}%")
print("from this i calculate the risk percentage according to loan_status")
```

```python
import pandas as pd
df=pd.read_csv("D:\\Internship\\loan\\loan.csv")
df['expected_amount'] = df['loan_amnt'] + df['total_rec_int']
total_expected = df['expected_amount'].sum()
total_received = df['total_pymnt'].sum()
difference = total_received - total_expected
profit_percentage = (difference / total_expected) * 100
print(f"Total Expected Amount: {total_expected}")
print(f"Total Received Amount: {total_received}")
print(f"Difference: {difference}")
print(f"Profit Percentage: {profit_percentage:.2f}%")
print("From this we calculate the profit of bank according to the monry given by the bank and taken back with intrest")
```
